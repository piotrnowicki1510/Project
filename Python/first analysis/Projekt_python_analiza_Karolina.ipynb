```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
e_df = pd.read_csv('C:/Users/karol/OneDrive/Pulpit/IBM.csv')
```

```python
e_df
```

```python
e_df['Attrition'] = e_df['Attrition'].apply(lambda x:1 if x == 'Yes' else 0)
```

```python
e_df
```

```python
# How many employees left the company
left_df = e_df[e_df['Attrition'] == 1]
# How many employees stayed in the company
stayed_df = e_df[e_df['Attrition'] == 0]
```

```python
print(len(e_df))
print(len(left_df))
print(len(stayed_df))
```

```python
plt.figure(figsize = (5,5))
sns.countplot(x = 'NumCompaniesWorked', hue = 'Attrition', data = e_df)
pass
```
