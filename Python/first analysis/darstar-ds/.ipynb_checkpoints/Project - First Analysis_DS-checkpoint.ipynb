```python
df_ds_TTLY_YES = df_ds.loc[df_ds["Attrition"]=="Yes", ["Attrition", "TrainingTimesLastYear"]]
# df_ds_TTLY_YES.sort_values("TrainingTimesLastYear")
print(type(df_ds_TTLY_YES))
df_ds_TTLY_YES.value_counts()
# df_ds_TTLY_YES
```

```python
# Liczba pracowników, którzy odeszli i liczba szkoleń w ostatnim roku
df_ds_TTLY_YES = df_ds[df_ds["Attrition"]=="Yes"]
# print(type(df_ds_TTLY_YES))
df_ds_TTLY_YES_only = df_ds_TTLY_YES.groupby(by="TrainingTimesLastYear").count() ["Attrition"].reset_index()
# df_ds_TTLY_YES_only2 = pd.DataFrame(df_ds_TTLY_YES_only)
df_ds_TTLY_YES_only = df_ds_TTLY_YES_only.rename(columns = {"Attrition":"Attrition_YES"})
# df_ds_TTLY_YES.sort_values("TrainingTimesLastYear")
# df_ds_TTLY_NO_only = df_ds_TTLY_NO["TrainingTimesLastYear"].value_counts()
# df_ds_TTLY_NO.to_frame()
# df_ds_TTLY_NO
print(type(df_ds_TTLY_YES_only))
df_ds_TTLY_YES_only
```

```python
# Liczba pracowników, którzy zostali i liczba szkoleń w ostatnim roku
df_ds_TTLY_NO = df_ds[df_ds["Attrition"]=="No"]
# print(type(df_ds_TTLY_NO))
# print(df_ds_TTLY_NO[["TrainingTimesLastYear", "Attrition"]])
df_ds_TTLY_NO_only = df_ds_TTLY_NO.groupby(by="TrainingTimesLastYear").count() ["Attrition"].reset_index()
print(df_ds_TTLY_NO_only)
# df_ds_TTLY_NO_only2 = pd.DataFrame(df_ds_TTLY_NO_only)
# print(df_ds_TTLY_NO_only2)
# print(df_ds_TTLY_NO_only2.columns)
df_ds_TTLY_NO_only = df_ds_TTLY_NO_only.rename(columns = {"Attrition":"Attrition_NO"})
# df_ds_TTLY_YES.sort_values("TrainingTimesLastYear")
# df_ds_TTLY_NO_only = df_ds_TTLY_NO["TrainingTimesLastYear"].value_counts()
# df_ds_TTLY_NO.to_frame()
# df_ds_TTLY_NO
print(type(df_ds_TTLY_NO_only))
df_ds_TTLY_NO_only
```

```python
df_ds_TTLY_Y_N = pd.merge(df_ds_TTLY_YES_only, df_ds_TTLY_NO_only, on="TrainingTimesLastYear", how="left")
df_ds_TTLY_Y_N.reset_index()
df_ds_TTLY_Y_N["Yes+No"] = df_ds_TTLY_Y_N["Attrition_YES"] + df_ds_TTLY_Y_N["Attrition_NO"]
df_ds_TTLY_Y_N["Probability_YES"] = df_ds_TTLY_Y_N["Attrition_YES"]/df_ds_TTLY_Y_N["Yes+No"]
df_ds_TTLY_Y_N["Cum_Yes+No"] = df_ds_TTLY_Y_N["Yes+No"].cumsum()
df_ds_TTLY_Y_N["Cum_Attr_Yes"] = df_ds_TTLY_Y_N["Attrition_YES"].cumsum()
df_ds_TTLY_Y_N["Cum_Prob_Yes"] = df_ds_TTLY_Y_N["Cum_Attr_Yes"]/df_ds_TTLY_Y_N["Cum_Yes+No"]
print(df_ds_TTLY_Y_N)
print(type(df_ds_TTLY_Y_N))
# df_ds_TTLY_Y_N.info()
# print(df_ds_TTLY_Y_N.loc["TrainingTimesLastYear" == 2])
```

```python
from IPython.display import display
from ipywidgets import HBox, Label

TTLY_slider = widgets.IntSlider(
    value=0,
    min=0,
    max=6,
    step=1,
    description='',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

# display(TTLY_slider)
HBox([Label('Liczba szkoleń w ostatnim roku:'), TTLY_slider])
```

```python
def value_handler(v):
    TrainNo = TTLY_slider.value
    Attr_Y_prob = df_ds_TTLY_Y_N.loc[df_ds_TTLY_Y_N["TrainingTimesLastYear"]==TrainNo, "Cum_Prob_Yes"]
    print("Prawdopodobieństwo odejścia z firmy wynosi: {}".format(Attr_Y_prob[TrainNo]))


widgets.interact(value_handler, v=TTLY_slider)
```
