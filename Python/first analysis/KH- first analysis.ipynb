```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

df
```

```python
df.columns
```

```python
df[df.columns[:8].unique()] # analized columns
df['Attrition'].replace({"Yes":1,"No":0},inplace=True) #zmiana na wartosci 1/0 - 1=YES, 0 =NO
# 1 - left job, 0 - stayed
```

```python
df_kh = df[df.columns[:8].unique()]

df_kh
```

```python
df_kh.describe()
```

```python
YES = df_kh[df_kh['Attrition'] == 1]

NO = df_kh[df_kh['Attrition'] == 0]

print(len(df_kh))  #quantity all
print(len(YES))  #quantity - people who left
print(len(NO))  #quantity - people who stayed
```

## ATTRITION - YES

Employees who left the company

```python
df_yes = df_kh.loc[df_kh['Attrition']==1]
df_yes
```

```python
df_yes.describe()
```

```python
grby_dep = df_yes.groupby(by = "Department")["Department","DailyRate", "Education"] #group by department
print(grby_dep.mean())
print(grby_dep.max())
print(grby_dep.min())
print(grby_dep.count())
```

## ATTRITION - NO

Employees who stayed at the company

```python
df_no = df_kh.loc[df_kh['Attrition']==0]
df_no
```

```python
df_no.describe()
```

```python
grby_dep2 = df_no.groupby(by = "Department")["Department","DailyRate", "Education"] #group by department
print(grby_dep2.mean())
print(grby_dep2.max())
print(grby_dep2.min())
print(grby_dep2.count())
```
